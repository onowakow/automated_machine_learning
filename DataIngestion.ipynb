```python
import pandas as pd
import requests

abalone_dataset_url = (
    "http://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
)
abalone_columns = [
    "sex",
    "length",
    "diameter",
    "height",
    "whole_weight",
    "shucked_weight",
    "viscera_weight",
    "shell_weight",
    "rings",
]
abalone_data = pd.read_csv(abalone_dataset_url, names=abalone_columns)
```
